```python
import os
import pandas as pd
import numpy as np
import math
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib import cm
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from dateutil import parser
import io
import base64
from IPython.display import HTML
from imblearn.under_sampling import RandomUnderSampler
```

```python
# カテゴリ変数のパース(仮実装で、値に変換するだけ)
# 最終的にはビットマップっぽいので、エンコーディングする。
from parse import *
def cat_data_to_val(x):
    if type(x) is str:
        r = parse("T{}", x)
        return float(r[0])
    else:
        return x
```

```python

def is_bit_on(s, pos):
    if type(s) is str:
        r = parse("T{}", s)
        val = (int(r[0]))
        mask = 1 << pos
        if val & mask:
            return 1
        else:
            return 0
    else:
        return 0
```

```python
def decode_categorical_data(df):
    for column_name, item in df.iteritems():
        if column_name != 'Id':
            vallist = item.unique()
            bitlist = []
            for c in vallist:
                if type(c) is str:
                    r = parse("T{}", c)
                    val = int(r[0])
                    if val < 0:
                        val = val + 4294967296
                    bitpos = int(np.log2(val))
                    bitlist.append(bitpos)

            for bit in bitlist:
                df[column_name + "_bit_" + str(bit)] = df[column_name].apply(lambda x:is_bit_on(x, bit)).astype('int8')
            df.drop(column_name, axis=1, inplace=True)

    
```

```python
import xgboost as xgb


def separate_X_y(df):
    X = df.drop(['Response'], axis=1)
    y = df['Response']
    return X, y

 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

def train_with_r_forest(df):
    X_train, y_train = separate_X_y(df)

    cl_weight = {0:0.3, 1:0.7}
    
    rf = RandomForestClassifier(max_depth=5,n_estimators=10)
    rf.fit(X_train, y_train)
    return rf



```

 
OKデータをアンダーサンプリングする<br>
カテゴリカルデータも対応するIDだけ残す

```python
def train_and_predict_one_chnuk(df_num, df_cat, df_result, nrows):
    df_num_ok   = df_num[df_num['Response'] == 0]
    df_num_ng   = df_num[df_num['Response'] == 1]
    df_num_test = df_num[df_num['Response'] == -1]

    undersample_rate = len(df_num_ng) * 5 / len(df_num_ok)
    df_num_ok_sample = df_num_ok.sample(frac = undersample_rate)

    df_num_balance = pd.concat([df_num_ok_sample, df_num_ng, df_num_test])
    print('Response = 0のテーブル', df_num_ok.shape)
    print('Response = 1のテーブル', df_num_ng.shape)
    print('Response = 0をアンダーサンプルした後のフレーム', df_num_balance.shape)
    
    # カテゴリカルデータも対応するIDだけ残す。
    df_ids = df_num_balance.loc[:, ['Id']]
    df_cat_balance = pd.merge(df_ids, df_cat, on=['Id'], how = 'left')
    
    # 全部 NaN な列をdropして、Nanを大きな値で置換。
    print("before" + str(df_num.shape))
    df_num_balance.dropna(how='all', axis=1, inplace=True)
    df_num_balance.fillna(999, inplace=True)
    print("after" + str(df_num.shape))
    
    decode_categorical_data(df_cat_balance)
    
    df =  pd.merge(df_cat_balance, df_num_balance, on=['Id'], how = 'left')
    df_train = df[(df['Response'] == 0) | (df['Response'] == 1)] 
    df_test  = df[df['Response'] == -1]

    rf = train_with_r_forest(df_train)

    X_test, y_test = separate_X_y(df_test)
    y_pred = rf.predict(X_test)
    
    df_result_add = pd.DataFrame(columns=['Id', 'Response'])
    df_result_add.loc[:, 'Id']       = df_test['Id'].values
    df_result_add.loc[:, 'Response'] = y_pred
    df_result = pd.concat([df_result, df_result_add])
    
    return df_result
    
```

```python
CHUNKSIZE = 50000
#CHUNKSIZE = 1000
NROWS = 1200000
#NROWS = 1500

chunk_num_train = pd.read_csv("../../data/train_numeric.csv",     chunksize = CHUNKSIZE)
chunk_cat_train = pd.read_csv("../../data/train_categorical.csv", chunksize = CHUNKSIZE)
chunk_num_test = pd.read_csv("../../data/test_numeric.csv",     chunksize = CHUNKSIZE)
chunk_cat_test = pd.read_csv("../../data/test_categorical.csv", chunksize = CHUNKSIZE)

df_result = pd.DataFrame(columns=['Id', 'Response'])
nrows = 0
while True:
    df_num_train = chunk_num_train.get_chunk(CHUNKSIZE)
    df_cat_train = chunk_cat_train.get_chunk(CHUNKSIZE)
    df_num_test  = chunk_num_test.get_chunk(CHUNKSIZE)
    df_cat_test  = chunk_cat_test.get_chunk(CHUNKSIZE)
    
    df_num_test['Response'] = -1
    
    df_num = pd.concat([df_num_train, df_num_test])
    df_cat = pd.concat([df_cat_train, df_cat_test])
    
    df_result = train_and_predict_one_chnuk(df_num, df_cat, df_result, nrows)
   
    nrows += CHUNKSIZE
    if nrows >= NROWS:
        break


```
